```python
import matplotlib as mpl
mpl.use('pgf')
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(0, 1, 48000)
k_a = .5

f_m = 2
m = np.cos(2 * np.pi * f_m * n)

A_c = 2 
f_c = 20
c = A_c * np.cos(2 * np.pi * f_c * n)

s = (1 + k_a * m)

y = c * s

fig, ax = plt.subplots(1, 3)
[axi.set_axis_off() for axi in ax.ravel()]
[axi.axhline(0, color = 'black', linewidth = .75) for axi in ax.ravel()]
[axi.axvline(0, color = 'black', linewidth = .75) for axi in ax.ravel()]
[axi.set_ylim(bottom=-3.2, top=3.2) for axi in ax.ravel()]
# [axi.grid(True, which='both') for axi in ax.ravel()]

(ax1, ax2, ax3) = ax
# fig.suptitle('Horizontally stacked subplots')
fig.set_figheight(8)
fig.set_figwidth(18)
ax1.plot(s, color = 'black')
ax1.text(1.1, 0.5, r'$\times$', horizontalalignment='center', verticalalignment='center', transform=ax1.transAxes, fontsize = '44')
# ax1.text(0.5,-0.1, r'$[1+k_a m(t)]', size=12, ha="center", transform=ax1.transAxes)
ax1.text(0.5,1, r'$[1+k_a m(t)]', size=24, ha="center", transform=ax1.transAxes)


ax2.plot(c, color = 'black')
ax2.text(1.1, 0.5, '=', horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes, fontsize = '44')
ax2.text(0.5,1, r'$c(t)$', size=24, ha="center", transform=ax2.transAxes)


ax3.plot(y, color = 'black')
ax3.text(0.5,1, r'$A_c[1+k_a m(t)]\cos(2\pi f_c t)$', size=24, ha="center", transform=ax3.transAxes)
plt.savefig('chap_01_example.pgf', format='pgf')
plt.show()
```
