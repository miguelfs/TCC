### Demodulacao incoerente

```python
def demod_hilb(x):
    m = np.abs(x)
    c = np.cos(np.angle(x))
    return (m,c)

def low_pass(x, fs, order = 10, cutoff_freq = 32) -> y:
    sos = signal.butter(order, cutoff_freq, 'lp', fs=fs, output='sos')
    return signal.sosfilt(sos, x)

def low_pass_coefs(fs, order = 10, cutoff_freq = 32):
    b, a = signal.butter(order, cutoff_freq, fs=fs)
    return signal.findfreqs(b, a, order)
```

```python
m,c = demod_hilb(x)
m_filtered = low_pass(m, fs)
output = m_filtered * c

# Plot:
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].plot(n, m_filtered)
axes[0].set_title('Filtered modulation')
axes[0].set_ylabel('m\'[n]')
axes[0].set_xlabel('n (samples)')

axes[1].plot(n, output)
axes[1].set_title('Filtered modulation with carrier')
axes[1].set_ylabel('x\'[n]')
axes[1].set_xlabel('n (samples)')


plt.tight_layout()
plt.show()
```

```python
def hard_low_pass(order = 64, cutoff = 10):
    filter = [1 for _ in range(order)]
    return filter[:cutoff]+[0]*(order-cutoff)

def hard_high_pass(order = 64, cutoff=6):
    filter = [0 for _ in range(order)]
    return filter[:cutoff]+[1]*(order-cutoff)

def as_int_16 (array: list):
    b = 16
    assert((max(array) <= 1) and (min(array) >= -1))
    return [v * (2**(b-1) - 1) for v in array]

def as_PCM (array: list):
    b = 16
    assert((max(array) <= 32767) and (min(array) >= -32767))
    return [v / (2**(b-1) - 1) for v in array]
```

```python
# Será uma estrutura com decimação?
    # Os filtros tem que ser complementares?
# O espaçamento é linear? geométrico? em oitavas?
    # Qual o espaçamento dos filtros? Precisa de fase linear?
    
# sugestão: usar filtros modulados por cosseno (cossine modulated filters),
# 

def PQMF_from_file(input_file = 'samples/starwars.wav', output_file = 'output.wav', order = 64, filtro = [1 for _ in range(64)]) :
    N = order
    # QMF window coefficients
    qmfwin=np.loadtxt('qmf.dat');
    # Analysis Folding matrix:
    Fa = ha2Fa3d_fast(qmfwin, N)
    # Synthesis Folding matrix:
    Fs = hs2Fs3d_fast(qmfwin, N)
    # read original file, length and sample rate
    wf = wave.open(input_file, 'rb')
    length = wf.getnframes()
    sr = wf.getframerate()

    # open sound file to write the reconstruced sound:
    wfw = wave.open(output_file, 'wb')
    wfw.setnchannels(1)
    # 2 bytes of width is equivalent to 16 bits
    wfw.setsampwidth(2)
    # set frame rate
    wfw.setframerate(sr)
    pqmf = PQMF()

    for m in range(int(length / N)):
        # Analysis:
        data = wf.readframes(N)
        x = list(struct.unpack('h' * N, data))
        y = pqmf.analysisqmf_realtime(x, Fa, N)
        if (filtro is not None):
            y = y * filtro

        # Synthesis:
        xrek = pqmf.synthesisqmf_realtime(y, Fs, N)
        # write with 2 Bytes per sample:
        # Convert to short ints:
        xrek = np.array(xrek, dtype='int16')
        data = struct.pack('h' * N, *xrek)
        wfw.writeframes(data)
    wf.close()
    wfw.close()
    
def PQMF_from_array(input, fs, output_file = 'som.wav', order = 64, filtro = [1 for _ in range(64)]):    
    N = order
    # QMF window coefficients
    qmfwin=np.loadtxt('qmf.dat');
    # Analysis Folding matrix:
    Fa = ha2Fa3d_fast(qmfwin, N)
    # Synthesis Folding matrix:
    Fs = hs2Fs3d_fast(qmfwin, N)
    length = len(input)
    input = as_int_16(input)
    output = [0 for _ in range(length)]

    # open sound file to write the reconstruced sound:
    wfw = wave.open('som.wav', 'wb')
    wfw.setnchannels(1)
    wfw.setsampwidth(2)
    wfw.setframerate(fs)
    
    pqmf = PQMF()

    for m in range(int(length / N)):
        # Analysis:
        x = input[m*N:(m+1)*N]
        y = pqmf.analysisqmf_realtime(x, Fa, N)
        y = y * filtro

        # Synthesis:
        xrek = pqmf.synthesisqmf_realtime(y, Fs, N)
        # write with 2 Bytes per sample:
        # Convert to short ints:
        xrek = np.array(xrek, dtype='int16')
        data = struct.pack('h' * N, *xrek)
        wfw.writeframes(data)
        output[m*N:(m+1)*N] = xrek

#     wf.close()
    wfw.close()
    return as_PCM(output)
```
