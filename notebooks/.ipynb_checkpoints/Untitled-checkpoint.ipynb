```python
import numpy as np
import librosa
from IPython.display import Audio
from matplotlib import pyplot as plt
plt.style.use('dark_background')
```

### 1) Time-domain:

```python
filename = librosa.ex('trumpet')
y, sr = librosa.load(filename)
Audio(y, rate = sr, autoplay=True)
```

```python
plt.figure(figsize=(16, 6))
plt.plot(y)
plt.show()

plt.figure(figsize=(16, 6))
plt.plot(y)
plt.xlim(44000, 49000)
plt.show()

M = 1024
plt.figure(figsize=(16, 6))
plt.plot(y)
plt.xlim(45000, 45000 + M)
plt.show()
```

```python
y, sr = librosa.load(filename)
Audio(y[44000:49000], rate = sr, autoplay=True)
```
